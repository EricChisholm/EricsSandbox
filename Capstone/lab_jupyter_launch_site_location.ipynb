```python
## Task 1: Mark all launch sites on a map
# === Task 1: Mark all SpaceX launch sites on a Folium map ===

import piplite
# Install all needed deps up front (safe to re-run)
await piplite.install([
    'jinja2',      # templating for branca/folium
    'requests',    # HTTP utils used by folium
    'branca',      # folium base
    'xyzservices', # tile provider registry
    'folium',
    'pandas',
])

import folium
from folium.plugins import MousePosition, MarkerCluster
from folium.features import DivIcon

# Known SpaceX launch site coordinates
launch_sites = {
    "CCAFS SLC 40": {"lat": 28.562302, "lon": -80.577356, "label": "Cape Canaveral SLC-40 (Florida)"},
    "KSC LC 39A":  {"lat": 28.573255, "lon": -80.646895, "label": "Kennedy Space Center LC-39A (Florida)"},
    "VAFB SLC 4E": {"lat": 34.632093, "lon": -120.610829, "label": "Vandenberg SLC-4E (California)"},
    # Add more pads here later if needed (e.g., Starbase/Boca Chica)
}

# Center map at mean lat/lon of sites
mean_lat = sum(s["lat"] for s in launch_sites.values()) / len(launch_sites)
mean_lon = sum(s["lon"] for s in launch_sites.values()) / len(launch_sites)

m = folium.Map(location=[mean_lat, mean_lon], zoom_start=5, control_scale=True, tiles="OpenStreetMap")

# Mouse position readout
MousePosition(
    position="bottomleft",
    prefix="Lat/Lon:",
    separator=" | ",
    num_digits=6
).add_to(m)

# Cluster (handy if you add more sites)
cluster = MarkerCluster(name="Launch Sites").add_to(m)

# Markers + small labels
for code, info in launch_sites.items():
    folium.Marker(
        location=[info["lat"], info["lon"]],
        popup=f"<b>{code}</b><br>{info['label']}",
        tooltip=code,
        icon=folium.Icon(icon="rocket", prefix="fa")
    ).add_to(cluster)

    folium.Marker(
        [info["lat"], info["lon"]],
        icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(0, 0),
            html=f'<div style="font-size:12px;font-weight:600;background:rgba(255,255,255,0.85);padding:2px 4px;border-radius:4px;">{code}</div>',
        )
    ).add_to(m)

folium.LayerControl(collapsed=False).add_to(m)

# Display the map (inline)
m

# Save to file for later reference / download (optional)
m.save("spacex_launch_sites_map.html")
```

```python
# Task 2: Mark the success/failed launches for each site on the map
# Ensure spacex_df and launch_sites_df already exist as in Task 1:
# spacex_df columns used: ['Launch Site','Lat','Long','class']

import folium
from folium.plugins import MarkerCluster
from folium.features import DivIcon

# Initialize a new map centered between the sites
mean_lat = launch_sites_df['Lat'].mean()
mean_lon = launch_sites_df['Long'].mean()
result_map = folium.Map(location=[mean_lat, mean_lon], zoom_start=5, control_scale=True)

# Optional: cluster to keep the map tidy
cluster = MarkerCluster(name="All Launches").add_to(result_map)

# Color by launch outcome
color_map = {1: 'green', 0: 'red'}

# Add a small circle marker for every launch
for _, r in spacex_df.iterrows():
    outcome = 'Success' if r['class'] == 1 else 'Failure'
    folium.CircleMarker(
        location=[r['Lat'], r['Long']],
        radius=5,
        color=color_map[r['class']],
        fill=True,
        fill_opacity=0.7,
        tooltip=f"{r['Launch Site']} — {outcome}",
        popup=f"<b>{r['Launch Site']}</b><br>Outcome: {outcome}"
    ).add_to(cluster)

# Add one larger labeled circle per site (helps orientation)
for _, s in launch_sites_df.iterrows():
    coord = [s['Lat'], s['Long']]
    label = s['Launch Site']
    folium.Circle(coord, radius=1000, color='#000000', fill=False).add_to(result_map)
    folium.Marker(
        coord,
        icon=DivIcon(
            icon_size=(160, 24),
            icon_anchor=(0, 0),
            html=f'<div style="font-size:12px;font-weight:600;background:rgba(255,255,255,0.85);padding:2px 4px;border-radius:4px;">{label}</div>'
        )
    ).add_to(result_map)

folium.LayerControl(collapsed=False).add_to(result_map)

# Display the interactive map
result_map

# (Optional) Save for later viewing/downloading
# result_map.save("spacex_task2_success_failure_map.html")
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

# Haversine function to calculate distance (km)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in km
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
