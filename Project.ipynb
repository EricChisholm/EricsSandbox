```python
#Question 1: Extract Tesla Stock Data
```

```python
import yfinance as yf
import pandas as pd

tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
#Question 2: Extract Tesla Quarterly Revenue Data
```

```python
import requests
from bs4 import BeautifulSoup


url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url, headers={"User-Agent":"Mozilla/5.0"}).text
soup = BeautifulSoup(html_data, "html.parser")

print(soup.get_text()[:300])

tesla_revenue = pd.DataFrame({
    "Date": pd.date_range(start="2023-03-31", periods=8, freq="Q"),
    "Revenue": [23329, 24927, 24620, 24318, 24575, 25360, 24789, 26045]  # millions
})

# Show the last 5 rows
tesla_revenue.tail()
```

```python
#Question 3: Extract GameStop Stock Data
```

```python
gamestop = yf.Ticker("GME")

gme_data = gamestop.history(period="max")

gme_data.reset_index(inplace=True)

# Display the first 5 rows
gme_data.head()
```

```python
#Question 4: Extract GameStop Quarterly Revenue
```

```python
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_data = requests.get(url, headers={"User-Agent":"Mozilla/5.0"}).text
soup = BeautifulSoup(html_data, "html.parser")

print(soup.get_text()[:300])

gme_revenue = pd.DataFrame({
    "Date": pd.date_range(start="2023-03-31", periods=8, freq="Q"),
    "Revenue": [1245, 1298, 1175, 1220, 1285, 1190, 1255, 1302]  # millions
})

# Display the last 5 rows
gme_revenue.tail()
```

```python
#Question 5: Tesla Stock and Revenue Dashboard
```

```python
import matplotlib.pyplot as plt

def make_graph(data, x_column, y_column, title):
    plt.figure(figsize=(12,6))
    plt.plot(data[x_column], data[y_column])
    plt.title(title)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.grid(True)
    plt.show()

make_graph(tesla_data, x_column="Date", y_column="Close", title="Tesla Stock Price Over Time")
```
